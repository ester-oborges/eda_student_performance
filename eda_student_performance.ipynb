## **1.1. Mesclagem de Datasets**

## **2.2. Sumarização numérica**

### **2.2.1. Análise da tabela de resumo numérico (variáveis discretas)**

```python
def resumo_sum(df, colunas):
  return pd.DataFrame({
      'Média': df[colunas].mean(),
      'Mediana': df[colunas].median(),
      'Moda': df[colunas].mode().iloc[0],
      'IQR': df[colunas].quantile(0.75) - df[colunas].quantile(0.25)
  })

sum_d3 = resumo_sum(d3, discretas)

sum_d3
```

### **2.2.2. Análise das tabelas de frequência (variáveis nominais e ordinais)**

```python
def resumo_freq(df, colunas):
  todas = []
  for col in colunas:
    freq_abs = df[col].value_counts()
    freq_rel = df[col].value_counts(normalize=True).round(3) * 100
    tabela = pd.DataFrame({
        "Categoria": freq_abs.index,
        "Frequência absoluta": freq_abs.values,
        "Frequência relativa (%)": freq_rel
    })
    tabela["Variável"] = col
    todas.append(tabela)
  return pd.concat(todas, ignore_index=True)

freq_d3 = resumo_freq(d3, nominais)

freq_d3
```

## **4.2. Principais insights**

**Diferença disciplinar:**
- **Matemática:** apresenta maior incidência de reprovações (+30%) e faltas (+10%) e uma queda de desempenho entre a 1ª e 3ª etapas (`G1` → `G3`). É a disciplina de maior dificuldade, independentemente da faixa etária.

**Preditores de desempenho (Fatores-Chave)**
- **Motivação e aspiração**: alunos que pretendem cursar o ensino superior (`higher`) e que escolhem a escola pela qualidade do curso (`reason`) demonstram as melhores médias.
- **Contexto sociocultural**: maior escolaridade dos pais (mães em `services` ou `at_home`, pais em `service`s → `Mjob`, `Fjob`) e residência na zona urbana (`address` → `U`) estão associados a um desempenho consistentemente superior.
- **Fatores institucionais:** a escola `GP` tem médias significativamente mais altas, sugerindo diferenças de qualidade pedagógica ou infraestrutura.
<br>

**Efeitos compensatórios/seleção inversa**
- Variáveis como suporte escolar (`schoolsup`), suporte familiar (`famsup`) e aulas extras pagas (`paid`) estão associadas a um desempenho inferior. Isso sugere que esses recursos são procurados ou fornecidos a alunos que já estão em situação de vulnerabilidade acadêmica, atuando como um efeito compensatório e não como uma causa de melhora.
<br>

**Consistência de comportamento**

Há uma forte correlação entre notas (`G1`, `G2`, `G3`) e uma correlação moderada entre o desempenho de Matemática e Português. Alunos com mais faltas ou reprovações em uma disciplina tendem a repetir esse padrão na outra, indicando um estilo de engajamento consistente (ou falta dele).
